2) Getting Useful Information from the Dataset

```
# This function builds a dataframe with all the attribute's counted
def valueCounts():
  columnNames = []
  valueCounts = pd.DataFrame(columns = columnNames)
  for c in df.columns:
      values = df[c].value_counts()
      valueCounts = valueCounts.append(values)
  valueCounts.columns = ['Legit','Phish','Suspect']
  return valueCounts
# This function builds a dataframe with attributes with suspicious values
def valueCountsSuspect():
  valueCount = pd.DataFrame()
  valueCount = valueCounts()
  counts_with_suspect = valueCount.dropna()
  counts_with_suspect_sorted = counts_with_suspect.sort_values('Suspect' , ascending=False)
  return counts_with_suspect_sorted

# This function builds a datafram showing the distribution of the attributes
def valueCountsDistributions():
  valueCount = pd.DataFrame()
  valueCount = valueCounts()
  valueCount['Diff'] = valueCount['Legit'] - valueCount['Phish']
  valueCountsSorted = valueCount.sort_values('Diff')
  return valueCountsSorted
```

```
# These are the Attributes that highly favor Phishing Emails
print(valueCountsDistributions().head())
```

```
# These are the Attributes that highly favor Legitmate Emails
print(valueCountsDistributions().tail())
```

```
# This Displays information about the Result Column of the dataset
def resultColumnInformation():
  valueCount = pd.DataFrame()
  valueCount = valueCounts()
  num_legit_result = valueCount.at['Result','Legit']
  num_phish_result = valueCount.at['Result','Phish']
  total = num_legit_result + num_phish_result
  percent_legit = round((num_legit_result / total) * 100,2)
  percent_phish = round((num_phish_result / total) * 100,2)
  print("Total Number of Data Rows:", total)
  print("Legitmate Results: " , num_legit_result , ":" , percent_legit , "%")
  print("Phishing Results: ", num_phish_result , ":" , percent_phish , "%")

resultColumnInformation()
```

 3) Cleaning the Class Labels and Inspecting for Missing Values

```
# This copys our data into a new dataframe, then replaces all 0s to NANs, then replaces NaNs with the mean
def imputeZerosToMean(): 
  replace_zero_df = df.copy() # Copy orginial dataframe
  replace_zero_df = replace_zero_df.replace(0,np.nan) # Replaces 0s with NaN
  df_mean_imputed = replace_zero_df.copy() #Copys that replaced dataframe 
  df_mean_imputed = df_mean_imputed.fillna(df_mean_imputed.mean()) #Imputes means into missing values
  return df_mean_imputed
```

```
# Display the Imputed Values
imputeZerosToMean().head()
```

```
# This function replaces all the -1s in the Result column with 0s.
# This is needed for machine learning algorithms to function
def formatResultsColumn():
  df_final = imputeZerosToMean().copy()
  df_final['Result'].replace({-1:0}, inplace = True)
  return df_final
```

```
# Displaying the formatted Result Column
df_final = formatResultsColumn()
df_final.Result
```

Splitting Data into Training and Testing Sets

```
# All Result Data
y_data = df_final.Result
# All Data without the Result column
x_data = df_final.drop('Result',axis=1)
#Splitting all the Data 80/20
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data ,test_size = 0.2, random_state=np.random.seed(8675309))
print('y_data ', y_data.shape)
print('x_data ', x_data.shape)
print('x_train ', x_train.shape)
print('x_test ', x_test.shape)
print('y_train ', y_train.shape)
print('y_test ', y_test.shape)
```

```
# Running a Logistic Regression with default parameters 
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
predictions = lr.predict(x_test)
score = lr.score(x_test, y_test)
print('---- Accuracy Score of Logistic Regression with default parameters ----')
print(score)

print('---- Classification Report of Logistic Regression with default parameters ----')
from sklearn.metrics import classification_report 
print(classification_report(y_test, predictions, target_names=['Phishing', 'Legitmate']))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```
# Defining variables for hyperparameters
penalty = ['l1', 'l2']
C = [0.8, 0.9 , 1.0]
tol = [0.01, 0.001, 0.001]
max_iter = [100, 150, 200 , 250]

# Savings all the hyperparameters into one dictionary
hyperparameters = { 'penalty' : penalty, 
                    'C' : C,
                    'tol' : tol,
                    'max_iter' : max_iter}
# Using a Randomized Search to determine the best hyperparameters 
from sklearn.model_selection import RandomizedSearchCV
randomSearchCV = RandomizedSearchCV(LogisticRegression(solver='liblinear'), param_distributions=hyperparameters)
randomSearchCV.fit(x_train,y_train)
best_score = randomSearchCV.best_score_
print('Best Score: ', best_score * 100, 'using' , randomSearchCV.best_params_)
```
